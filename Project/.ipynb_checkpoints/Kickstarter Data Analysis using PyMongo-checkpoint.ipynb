## Kicstarter Data Analysis
Kickstarter is the world's largest funding platform for creative projects. We analyze the Kickstarter project data in 2018 after loading it in MongoDB and then using PyMongo to fetch the data from the Mongo server to this notebook to perform further analysis.

We loaded our data in MongoDB in the below code:<br>
mongoimport -d Project -c KL_data --type CSV --file ks-projects.csv --headerline

```python
# Import necessary libraries
# Dataframe
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# Pymongo
import pymongo

# For pymongo aggregation
from bson.code import Code
```

```python
# Connect to our MongoDB database using PyMongo
from pymongo import MongoClient

client = MongoClient()
db = client['Project']
collection = db.KL_data
```

```python
# Verfiy if the data is properly imported from MongoDB
print(collection.find().count())
```

We have the same number of rows in the original data, therefore the data seems to be properly imported

### Get a bar plot of the top 10 category in terms of the number of kickstarter projects

```python
# Let's get the number of category count of succesfull projects
# Let's restrict the plotting to the top 10 categories
mapper = Code("""
            function(){
                emit(this.category,1);
            }
            """)
reducer = Code("""
            function(key,values){
                return Array.sum(values)
            }
""")

df = pd.DataFrame.from_records(collection.map_reduce(mapper,reducer,'test').find().sort('value',-1).limit(10))
```

```python
# Now after getting the data, plot it using matplotlib
plt.rcParams["figure.figsize"] = (16,5)
plt.bar(df['_id'],df['value'])
plt.title('Top 10 categories in terms of number of Kistarter projects')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()
```

### Get the distribution of the goals of all the different 'successful' projects

```python
df = pd.DataFrame.from_records(collection.find({"state":"successful"}))

plt.hist(df['goal'])
plt.title('Histogram of goals of successful Kicstarter projects')
plt.xlabel('Goals of projects in USD')
plt.ylabel('Count')
plt.show()
```

We can see that the goals of successful projects are extremely right skewed therfore visualizing the distribution after taking the log transformation

```python
import numpy as np
plt.hist(np.log(df['goal'].tolist()))
plt.title('Histogram of log goals of successful Kicstarter projects')
plt.xlabel('Log of goals of projects in USD')
plt.ylabel('Count')
plt.show()
```

We can see that this is much better now!

### Let's now get the top 10 projects who had the highest amount pledged by the crowd and were still live when this data was collected i.e in 2018

```python
df = pd.DataFrame.from_records(collection.find({'state':'live'}).sort('usd_pledged_real',-1).limit(10))
df
```

### Explore the relationship between the number of backers and the amount pledged in USD

```python
df = pd.DataFrame.from_records(collection.find())
```

```python
plt.scatter(df['backers'],df['usd_pledged_real'],label='Pledged amount in USD')
plt.plot(np.unique(df['backers']), np.poly1d(np.polyfit(df['backers'], df['usd_pledged_real'], 1))(np.unique(df['backers'])),'r',label='Best fit line')
plt.legend()
plt.xlabel('Number of backers')
plt.ylabel('Amount pledged in USD')
plt.title('Number of backers vs amount pledged in USD')
plt.show()
```

We can see that there is somewhat of a linear relationship between number of backers and the amount pledged however we can also see considerable number of outliers lying far from the line of best fit.

### Get the number of projects for the top 10 countries having most number of Kistarter projects

```python
mapper = Code("""
            function(){
                emit(this.country,1);
            }
            """)
reducer = Code("""
            function(key,values){
                return Array.sum(values)
            }
""")

df = pd.DataFrame.from_records(collection.map_reduce(mapper,reducer,'test').find().sort('value',-1).limit(10))
```

```python
# Plot the graph
plt.rcParams["figure.figsize"] = (10,10)
plt.pie(df['value'],labels=df['_id'])
plt.legend()
plt.title('Pie chart for the distribution of projects among the top 10 countries(in terms of the projects)')
plt.show()
```

```python
# Convert dates to proper formats
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
```

```python
# Get the amount of time when the project was launched on kickstart to its deadline
df['active_time'] = df['deadline'] - df['launched']
```

```python
# Get the days from the time difference
plt.hist(df['active_time'].dt.days)
plt.
plt.title('Distribution of days between launch and deadline')
```
